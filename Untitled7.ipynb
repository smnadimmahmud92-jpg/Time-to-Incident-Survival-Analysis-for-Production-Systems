```python
#Generate Survival Data

import os
import numpy as np
import pandas as pd

np.random.seed(123)
os.makedirs('data',exist_ok=True)
N_service=200
OBS_PER_SERVICE=5
rows=[]

for service_id in range(N_service):
  baseline=np.random.uniform(0.01,0.05)

  for _ in range(OBS_PER_SERVICE):
    cpu=np.random.uniform(0.2,0.95)
    mem=np.random.uniform(0.1,0.9)
    latency=np.random.uniform(50,600)
    error=np.random.uniform(0.0001,0.05)

    hazard_true=(
        baseline
        +0.8*cpu
        +0.6*mem
        +0.003*latency
        +10*error
    )
    T=np.random.exponential(scale=1/hazard_true)
    t=np.random.uniform(1,10)

    event=int(T<=t)
    time=min(T,t)

    rows.append([
        service_id,cpu,mem,latency,error,time,event
    ])

df=pd.DataFrame(rows,columns=[
    "service_id","cpu_load","memory_pressure","latency_ms","error_rate","duration","event"
])

df.to_csv("data/incident_survival.csv",index=False)
print("Save file",os.path.exists("data/incident_survival.csv"))
df.head()
```

```python
#Data sanity and censoring check

print(f"censoring rate:{(1-df.event.mean()):.2%}")
plt.hist(df[df.event==1]["duration"],bins=30,alpha=0.7,label="Incident")
plt.hist(df[df.event==0]["duration"],bins=30,alpha=0.7,label="censoring")
plt.legend()
plt.title("Observed Durations")
plt.xlabel("Time")
plt.ylabel("count")
plt.show()
```

```python
#Kaplan Meier Survival curves
kmf=KaplanMeierFitter()
plt.figure()
kmf.fit(df["duration"],df["event"])
kmf.plot_survival_function()
plt.title("Kaplan Meier Survival (Overall)")
plt.xlabel("Time")
plt.ylabel("Survival Probability")
plt.show()
```

```python
#KM by stress group
df["high_latency"]=(df["latency_ms"]>df["latency_ms"].median()).astype(int)
plt.figure()
for g, label in [(0,"Low Latency"),(1,"High latency")]:
  kmf.fit(
      df[df.high_latency==g]["duration"],
      df[df.high_latency==g]["event"],
      label=label
  )
  kmf.plot_survival_function()
  plt.title("Survival by latency Stress Group")
  plt.xlabel("Time")
  plt.ylabel("Survival Probability")
  plt.show()

```

```python
#Hazard Model and Likelihood

def hazard (beta,X):
  return np.exp(X @ beta)

def neg_log_likelihood(beta,X,t,event):
  lam=hazard(beta,X)
  ll=event*np.log(lam)-lam*t
  return -np.sum(ll)
```

```python
#Fit Model

X=df[["cpu_load","memory_pressure","latency_ms","error_rate"]].values
X=np.c_[np.ones(len(X)),X]
t=df["duration"].values
e=df["event"].values

res=minimize(
    neg_log_likelihood,
    np.zeros(X.shape[1]),
    args=(X,t,e),
    method="BFGS"
)
beta_hat=res.x
beta_hat
```

```python
#Train/Test evaluation
from sklearn.model_selection import train_test_split
X_tr,X_te,t_tr,t_te,e_tr,e_te=train_test_split(X,t,e,test_size=0.3,random_state=42)
res_tr=minimize(
    neg_log_likelihood,
    np.zeros(X.shape[1]),
    args=(X_tr,t_tr,e_tr),
    method="BFGS"
)
beta_tr=res_tr.x
risk_te=np.exp(X_te @ beta_tr)
c_index=concordance_index(t_te,-risk_te,e_te)
print("Test C-index",round(c_index,3))
```

```python
#Calibration by predicted risk groups
df["risk"]=np.exp(X @ beta_hat)
df["risk_group"]=pd.qcut(df["risk"],3,labels=["low","Medium","High"])
plt.figure()
for g in ["low","Medium","High"]:
  subset=df[df.risk_group==g]
  kmf.fit(subset["duration"],subset["event"],label=f"{g} risk")
  kmf.plot_survival_function()
plt.title("Calibration by Predicted Risk Group")
plt.xlabel("Time")
plt.ylabel("Survival Probability")
plt.show()
```

```python
#Cox Snell residual diagnostics
cum_hazard=np.exp(X @ beta_hat)*t
kmf.fit(cum_hazard,e)
plt.figure()
kmf.plot_survival_function(label="Empirical")
plt.plot(
    np.sort(cum_hazard),
    np.exp(-np.sort(cum_hazard)),
    "r--",
    label="Exp(1) reference"
)
plt.legend()
plt.title("Cox-Snell Residual Check")
plt.xlabel("Residual")
plt.ylabel("Survival")
plt.show()
```

```python
import numpy as np
import pandas as pd

feature_names = ["intercept","cpu_load","memory_pressure","latency_ms","error_rate"]
coef = pd.Series(beta_hat, index=feature_names)

# Hazard ratio per 1 unit increase (note: for latency and error_rate, better to use meaningful increments)
print("Coefficients:")
display(coef.to_frame("beta"))

# More meaningful hazard ratios:
increments = {
    "cpu_load": 0.10,          # +10% cpu
    "memory_pressure": 0.10,   # +10% memory pressure
    "latency_ms": 100,         # +100 ms
    "error_rate": 0.01         # +1% absolute error rate
}

hr = {}
for k,v in increments.items():
    hr[k] = float(np.exp(coef[k] * v))

pd.DataFrame({"increment": increments, "hazard_ratio": hr}).sort_values("hazard_ratio", ascending=False)
```
